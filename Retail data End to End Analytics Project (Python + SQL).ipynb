```python
df = pd.read_csv('orders.csv')
```

#### Handling null values


```python
df[df['Ship Mode']=='Not Available']
```

```python
df['Ship Mode'].unique()
```

```python
# Remove rows where 'Ship Mode' is either 'unknown' or NaN
df = df[~df['Ship Mode'].isin(['unknown','Not Available']) & df['Ship Mode'].notna()]

df['Ship Mode'].unique()
```

```python
df.head()
```

```python
#rename columns names
df.columns=df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

### Adding respective columns for discounts and profits

```python
df['discount']=df['list_price']*df['discount_percent']*.01
```

```python
df.head(2)
```

```python
df['sale_price'] = df['list_price']-df['discount']
```

```python
df.head(2)
```

```python
df['profit']=df['sale_price']-df['cost_price']
df.head(2)
```

```python
#converting order date from object data type to datetime
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
#dropping some columns I do not need
```

```python
df.drop(columns=['list_price','discount_percent','cost_price'],inplace=True)
```
